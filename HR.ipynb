# Data Analysis

1) Count of employees in each department

```python
count_emp=dataset.groupby('Department')['Department'].count()
```

```python
count_emp
```

```python
#pie graph
explode=[0,0,0.1,0,0,0]
palette_color=sns.color_palette('bright')
label=['Admin office','EO','IT','Pr','Sa','SE']
plt.pie(count_emp,autopct='%.0f%%',explode=explode,labels=label)
```

2) Average salary of employee in each department

```python
avg_salary=dataset.groupby('Department')['Salary'].mean()
```

```python
avg_salary.round(2)
```

```python
#pie graph
explode=[0,0,0.1,0,0,0]
palette_color=sns.color_palette('bright')
label=['Admin office','EO','IT','Pr','Sa','SE']
plt.pie(avg_salary,autopct='%.0f%%',explode=explode,labels=label)
```

3) Average salary of employee based on gender

```python
gen_salary=dataset.groupby(['Department','Sex'],as_index=False)['Salary'].mean()
```

```python
gen_salary
```

```python
sns.barplot(x='Department',y='Salary',hue='Sex',data=gen_salary)
```

4. Popular Recruitment Source of the company based on department.

```python
RS=dataset.groupby(['Department','RecruitmentSource']).size().reset_index(name='Counts')
```

```python
RS
```

```python
sns.barplot(x='RecruitmentSource',y='Counts',hue='Department',data=RS)
```
